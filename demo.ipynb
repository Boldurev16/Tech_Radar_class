Классификация сценариев по **quadrant** и **block**.

```python
from classifier import TechRadarClassifier
from evaluate import load_dataset, class_distribution

clf = TechRadarClassifier()
print('Dataset size:', len(load_dataset()))
print('Quadrant classes:', class_distribution(load_dataset())['quadrant'])
```

```python
metrics = clf.run_evaluate(holdout_ratio=0.2)
for key in [
    'quadrant_accuracy', 'quadrant_macro_f1',
    'block_accuracy', 'block_macro_f1', 'joint_accuracy', 'low_confidence_ratio'
]:
    print(f'{key}: {metrics[key]:.3f}')
```
